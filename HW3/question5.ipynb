```python
import cv2
import pylab
import numpy as np
from question4 import compute_disparity_new_mapped
from question1 import compute_disparity2_mapped
import time
```

```python
im1 = cv2.imread("subject/proj2-pair2-L.png", cv2.IMREAD_GRAYSCALE)
im2 = cv2.imread("subject/proj2-pair2-R.png", cv2.IMREAD_GRAYSCALE)

pylab.imshow(im1, cmap=pylab.gray())
pylab.show()
pylab.imshow(im2, cmap=pylab.gray())
pylab.show()

disparity = {}

win_size = 11
min_d = 10
max_d = 115
right = False
```

```python
im_id = 1
start = time.time()
disparity[im_id] = compute_disparity2_mapped(im1, im2,
                                         window_size=win_size,
                                         d_min=min_d,
                                         d_max=max_d,
                                         explore_right=right)
disparity[10+im_id] = compute_disparity2_mapped(im2, im1,
                                         window_size=win_size,
                                         d_min=min_d,
                                         d_max=max_d,
                                         explore_right=not right)
print(time.time() - start)
print(disparity[im_id].shape)
print(disparity[im_id].dtype)
print(np.max(disparity[im_id]))
print(np.min(disparity[im_id]))
pylab.imshow(disparity[im_id], cmap=pylab.gray())
pylab.show()
```

```python
im_id = 2
start = time.time()
disparity[im_id] = compute_disparity_new_mapped(im1, im2,
                                                window_size=win_size,
                                                d_min=min_d,
                                                d_max=max_d,                                            
                                                explore_right=right)
disparity[10+im_id] = compute_disparity_new_mapped(im2, im1,
                                                window_size=win_size,
                                                d_min=min_d,
                                                d_max=max_d,                                            
                                                explore_right=not right)
print(time.time() - start)
print(disparity[im_id].shape)
print(disparity[im_id].dtype)
print(np.max(disparity[im_id]))
print(np.min(disparity[im_id]))
pylab.imshow(disparity[im_id], cmap=pylab.gray())
pylab.show()
```

```python
mean_im1 = np.mean(im1)
mean_im2 = np.mean(im2)
print(mean_im1)
print(mean_im2)
print(mean_im2 / mean_im1)
median_im1 = np.median(im1)
median_im2 = np.median(im2)
print(median_im1)
print(median_im2)
print(median_im2 / median_im1)
print(np.min(im1), np.min(im2))
print(np.max(im1), np.max(im2))
```

```python
im1_cc = im1 * (np.mean(im2) / np.mean(im1))
```
